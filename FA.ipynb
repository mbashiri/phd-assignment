## Implementation

```python
def beta(lambd, psi):
    return np.dot(lambd.T, np.linalg.inv(psi + np.dot(lambd, lambd.T)))

def E_step(x, lambd, psi):
    """
    Returns the first and second moment of Z.

    Parameters
    ----------
    x : 2d array
        Observed data matrix
    lambd : 2d array
        factor loading matrix
    psi : 2d (square) array
        independent noise (U) covariance matrix
    
    Returns
    -------
    Ez : 2d array
        the expected value of the factors
    Ezz : 2d (square) array
        the expected covariance of the factors
    """
    b = beta(lambd, psi)
    Ez = np.dot(b, x)
    Ezz = np.eye(lambd.shape[1]) - np.dot(b, lambd) + np.dot(np.dot(b, x), np.dot(x.T, b.T)) 
    
    return Ez, Ezz

def M_step(x, Ez, Ezz):
    """
    Returns the updated lambda and psi
    
    Parameters
    ----------
    x : 2d array
        Observed data matrix
    Ez : 2d array
        the expected value of the factors
    Ezz : 2d (square) array
        the expected covariance of the factors
    
    Returns
    -------
    lambd_new : 2d array
        updated factor loading matrix with MLE
    psi_new : 2d (square) array
        updated independent noise (U) covariance matrix with MLE
    """
    lambd_new = np.dot(np.dot(x, Ez.T), np.linalg.inv(Ezz))
    psi_new = (1/x.shape[1]) * np.diag(np.diag(np.dot(x, x.T) - np.dot(lambd, np.dot(Ez, x.T))))
    
    return lambd_new, psi_new
```

However, in reality, we do not know the number of factors. So on a more fun part, I tried to use cross-validation and see if, using the implemented FA, we can guess the "correct" number of factors. This method used for this cross-validation (CV) is very much similar to using drop-out as a regularization method. We simply hold-out some of the entries of our matrix while doing the E-step and M-step. The held-out entries are essentially our "validation" set and the rest are out "training" set. Then we will end up having two reconstruction errors: (1) Training and (2) validation reconstruction errors. Trainining reconstruction error is only computed for the matrix entries which were not held-out for optimization and the validation reconstruction error is computed only for the heldout elements. Figure 4 shows the result of such cross-validation performed on both the data with low-level and high-level noise, to find the "best" number of factors.

## Generating synthetic data

We are going to assume 3 uncorrelated factors (i.e., neurons) with unit variance which give rise to 50 observed units (e.g., neurons).

```python
N, D, K_true = 200, 50, 3
```

### Factors (Z)

As mentioned in the paper, Z is assumed to be normally distributed with 0 mean and identity covariance matrix.

```python
mean = [0] * K_true
cov = np.eye(K_true)

np.random.seed(42)
Z = np.random.multivariate_normal(mean, cov, size=(N)).T
assert Z.shape == (K_true, N)
```

### Lambda

```python
np.random.seed(42)
lambd_true = np.random.normal(size=(D, K_true))
assert lambd_true.shape == (D, K_true)
```

### Psi and U (independent noise)

U is the independent noise in the data, that is given the factors there is (almost) zero correlation between each dimension of our observed data. Which follows that the covariance matrix (psi) of our noise to be a diagonal.

```python
# psi
np.random.seed(42)
psi_true = np.identity(D) * np.random.uniform(low=.2, high=.9, size=(D, D))
assert psi_true.shape == (D, D)

# U
mean = [0] * D
cov = psi_true
np.random.seed(42)
U_true = np.random.multivariate_normal(mean, cov, size=(N)).T
assert U_true.shape == (D, N)
```

### Observed data (X)

Now, we can compute the ground truth observed data. A constant is multiplied by `U` to let us vary the noise level (for the summary above, I used low-level = 1 and high-level = 5)

```python
X = np.dot(lambd_true, Z) + U_true * 1 # constant added to control the noise level
```

```python
fig = plt.figure(figsize=(7, 5)) #, dpi=300)

ax1 = fig.add_axes([0.1, 0.56, 0.43, 0.1])
ax2 = fig.add_axes([0.1, 0.43, 0.43, 0.1])
ax3 = fig.add_axes([0.1, 0.3, 0.43, 0.1])
axes1 = (ax1, ax2, ax3)

ax4 = fig.add_axes([0.5, 0.1, 0.3, 0.85])
ax5 = fig.add_axes([0.87, 0.1, 0.25, 0.85])
ax6 = fig.add_axes([1.25, 0.57, 0.35, 0.38])
ax7 = fig.add_axes([1.25, 0.1, 0.35, 0.38])

for ind, (z, ax) in enumerate(zip(Z, axes1)):
    sns.set_style("white")
    sns.despine(top=True, ax=ax)
    if ind != (K_true-1):
        sns.despine(bottom=True, ax=ax)
        ax.set_xticks([])
    else:
        ax.set_xlabel('Time (N)')
    ax.plot(z, lw=2)
    ax.set(xlim=(0, N), ylim=(-3, 3))

arr_img = plt.imread("Factors&Units.png", format='png')
ax4.imshow(arr_img)
sns.despine(ax=ax4, bottom=True, left=True)
ax4.set(xticks=[], yticks=[])

cbar_ax = fig.add_axes([1.13, .3, .03, .4])
sns.heatmap(X, cmap='bwr', ax=ax5, cbar=True, vmin=-8, vmax=8, cbar_ax=cbar_ax)
ax5.set(xlabel='Time (N)', ylabel='Units (D)')

sns.heatmap(lambd_true, ax=ax6, vmin=-3, vmax=3)
ax6.set(xlabel='Factor (K)', ylabel='Units (D)', xticks=[], yticks=[])

sns.heatmap(psi_true, ax=ax7, vmin=0, vmax=1)
ax7.set(xlabel='Units (D)', ylabel='Units (D)', xticks=[], yticks=[]);
```

## Computing factors and comparison with the ground truth

- Standardizing the data matrix 
- Defining the dimensions, as well as the "hypothesized" number of factors, which in this case is set to the ground truth just to check whether we can recover it or not
- Initializing lambda and psi

```python
data = X.copy()
data -= data.mean(axis=1).reshape(-1, 1)

(D, N), K = data.shape, K_true
```

```python
np.random.seed(42)
lambd = np.random.normal(size=(D, K)) * 1
psi = np.diag(np.diag(np.random.rand(D, D)))
```

Using the E-step and M-step above, now we can find the factors

```python
epochs = 10
for epoch in range(epochs):
    Ez, Ezz = E_step(data, lambd, psi)
    lambd, psi = M_step(data, Ez, Ezz)
    X_rec = np.dot(lambd, Ez)
    
    rec_err = np.mean(np.square(X - X_rec))
    print("Epoch {} | Reconstruction Error: {}".format(epoch, rec_err)) 
```

Now we are going to standardize our factors and also compute the reconstructed observed data - so we can also compute reconstruction error.

```python
Ez_scaler = StandardScaler().fit(Ez.T)
X_rec = np.dot(lambd, Ez)
```

```python
fig = plt.figure(figsize=(7, 5))

ax1 = fig.add_axes([0.1, 0.7, 0.43, 0.25])
ax2 = fig.add_axes([0.1, 0.4, 0.43, 0.25])
ax3 = fig.add_axes([0.1, 0.1, 0.43, 0.25])
axes1 = (ax1, ax2, ax3)

ax4 = fig.add_axes([0.6, 0.78, 0.27, 0.15])
ax5 = fig.add_axes([0.6, 0.55, 0.2, 0.2])
ax6 = fig.add_axes([0.6, 0.33, 0.27, 0.15])
ax7 = fig.add_axes([0.6, 0.1, 0.2, 0.2])
axes2 = (ax5, ax7)

ax8 = fig.add_axes([0.92, 0.1, 0.2, 0.85])
ax9 = fig.add_axes([1.14, 0.1, 0.2, 0.85])
axes3 = (ax8, ax9)

all_axes = (ax1, ax2, ax3, ax4, ax5, ax6, ax7, ax8, ax9)

for ind, (z, ez, ax) in enumerate(zip(Z, Ez_scaler.transform(Ez.T).T, axes1)):
    sns.set_style("white")
    sns.despine(top=True, ax=ax)
    if ind != (K_true-1):
        sns.despine(bottom=True, ax=ax)
        ax.set_xticks([])
    else:
        ax.set_xlabel('Time (N)')
    ax.plot(z)
    ax.plot(ez, 'r:')
    ax.set(xlim=(0, N), ylim=(-3, 3))

ax4.plot(lambd_true.flatten())
ax4.plot(lambd.flatten(), 'r:')
sns.despine(top=True, ax=ax4)
ax4.set(xticks=[])

ax6.plot(np.diag(psi_true))
ax6.plot(np.diag(psi), 'r:')
sns.despine(top=True, ax=ax6)
ax6.set(xticks=[])

sns.heatmap(lambd, ax=ax5, vmin=-3, vmax=3)
ax5.set(xlabel='Factor (K)', ylabel='Units (D)', xticks=[], yticks=[])

sns.heatmap(psi, ax=ax7, vmin=0, vmax=1)
ax7.set(xlabel='Units (D)', ylabel='Units (D)', xticks=[], yticks=[])

plot_these = [X_rec, np.square(X-X_rec)]
cbar_ax = fig.add_axes([1.37, .3, .03, .4])
ax_text = ['f', 'g']
for i, (x, ax) in enumerate(zip(plot_these, axes3)):
    sns.heatmap(x, cmap='bwr', ax=ax, cbar=i==0, vmin=-8, vmax=8, cbar_ax=None if i else cbar_ax)
    ax.set_xlabel('Time (N)')
    if i == 0:
        ax.set_ylabel('Units (D)')
        for tick in ax.get_yticklabels():
            tick.set_rotation(0)
    else:
        ax.set_yticks([])
```

## Cross-validation with FA

```python
np.random.seed(None)
N, D, K_true = 200, 50, np.random.randint(3, 20)
```

```python
# Z
mean = [0] * K_true
cov = np.eye(K_true)
np.random.seed(42)
Z = np.random.multivariate_normal(mean, cov, size=(N)).T
assert Z.shape == (K_true, N)

# lambda
np.random.seed(42)
lambd_true = np.random.normal(size=(D, K_true))
assert lambd_true.shape == (D, K_true)

# psi
np.random.seed(42)
psi_true = np.identity(D) * np.random.uniform(low=.2, high=.9, size=(D, D))
assert psi_true.shape == (D, D)

# U
mean = [0] * D
cov = psi_true
np.random.seed(42)
U_true = np.random.multivariate_normal(mean, cov, size=(N)).T
assert U_true.shape == (D, N)

X = np.dot(lambd_true, Z) + U_true * .5 # constant added to control the noise level
```

Choose a percentage of the entries in the data matrix to be ignored while optimizing the parameters:

```python
drop_perc = .3
np.random.seed(42)
M = np.random.rand(*data.shape) > np.random.uniform(size=data.shape) * drop_perc
```

Standardize and initialize the set the dimensions to initialize lambda and psi. We are also setting the maximum value for the range of K (number of factors) that we are planning to cross-validate across:

```python
data = X.copy()
data -= data.mean(axis=1).reshape(-1, 1)
```

```python
(D, N), K_max = data.shape, 30
```

```python
train_rec_errs, test_rec_errs = [], []
for K in np.arange(K_max):
    np.random.seed(42)
    lambd = np.random.normal(size=(D, K)) * 1
    psi = np.diag(np.diag(np.random.rand(D, D)))
    epochs = 30
    for epoch in range(epochs):
        Ez, Ezz = E_step(M * data, lambd, psi)
        lambd, psi = M_step(M * data, Ez, Ezz)
        data_rec = np.dot(lambd, Ez)
        train_rec_err = np.mean(np.square(M * (data - data_rec)))
        test_rec_err = np.mean(np.square((1 - M) * (data - data_rec)))
    print("K = {} | Train Reconstruction Error is {} | Test Reconstruction Error is {}".format(K, train_rec_err, test_rec_err))
    train_rec_errs.append(train_rec_err)
    test_rec_errs.append(test_rec_err)
```

```python
train_rec_errs = np.array(train_rec_errs)
test_rec_errs = np.array(test_rec_errs)
K_est = np.arange(K_max)[test_rec_errs == test_rec_errs.min()][0]

fig = plt.figure() #dpi=300)
plt.plot(np.arange(K_max), train_rec_errs / M[M==1].shape[0], 'bo', lw=5, label="Train Error")
plt.plot(np.arange(K_max), test_rec_errs / M[M==0].shape[0], 'ro', lw=5, label="Test Error")
plt.title("CVed K = {} | True K = {}".format(K_est, K_true), fontsize=20)
plt.xlabel("K values", fontsize=15)
plt.ylabel("Reconstruction Error", fontsize=15)
plt.xticks(fontsize=15)
plt.yticks(fontsize=15)
ax = plt.gca()
sns.despine(top=True, ax=ax)
```
